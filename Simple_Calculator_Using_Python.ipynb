```python
import math
def add(x,y):
    s=x+y  
    return s
def subtract(x,y):
    if (x>y):
        sub=x-y 
    else:
        sub=y-x
    return sub
def multiply(x,y):
    multi=x*y
    return multi
def divide(x,y):
    divi=x%y 
    return divi
def sqrt(x):
    sq=math.sqrt(x)
    return sq
def power(x,y):
    p=pow(x,y)
    return p
    print("Your answer is:%d"%(pow(x,y)))
def expo(x,y):
    ex=x**y  
    return ex
print("_____CALCULATOR_____")
while(True):
    print("ENTER THE OPERATION U NEED TO PERFORM.....")
    print("1.ADDITION")
    print("2.SUBTRACTION")
    print("3.MULTIPLICATION")
    print("4.DIVISION")
    print("5.SQUARE ROOT")
    print("6.POWER")
    print("7.EXPONENTIATION")
    print("8.EXIT")
    choice=int(input("Enter the choice: "))
    if (choice==1):
        x=int(input("Enter the value of x: ")) 
        y=int(input("Enter the value of y: "))
        s=add(x,y)
        print("Your answer is %d" %s)
    elif(choice==2):
        x=int(input("Enter the value of x: "))
        y=int(input("Enter the value of y: "))
        sub=subtract(x,y)
        print("Your answer is: %d" %sub)
    elif(choice==3):
        x=int(input("Enter the value of x: ")) 
        y=int(input("Enter the value of y: "))
        multi=multiply(x,y)
        print("Your answer is: %d" %multi)
    elif(choice==4):
        x=int(input("Enter the value of x: ")) 
        y=int(input("Enter the value of y: "))
        divi=divide(x,y)
        print("Your answer is: %d" %divi)
    elif(choice==5):
        x=int(input("Enter the value of x: "))
        y=int(input("Enter the value of y: "))
        sq=sqrt(x)
        print("Your answer is: %d" %sq)
    elif(choice==6):
        x=int(input("Enter the value of x: ")) 
        y=int(input("Enter the value of y: "))
        p=power(x,y)
        print("Your answer is: %d" %p)
    elif(choice==7):
        x=int(input("Enter the value of x: ")) 
        y=int(input("Enter the value of y: "))
        ex=expo(x,y)
        print("Your answer is: %d" %ex)
    elif(choice==8):
        print("YOU ARE EXIT....")
        break
    else:
        print("Enter a valid choice....")
```
